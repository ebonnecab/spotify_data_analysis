## Visualizing Data

#### Histogram of US Top 50 Audio Features

```python
#numerical data is separated into it's own dataframe
feats_df = us_top_df.select_dtypes(include = ['float64', 'int64'])
feats_df.head()
```

```python
#distribution plot using seaborn library
colu_names = feats_df.columns
fig, ax = plt.subplots(len(colu_names), figsize=(20,16))

#looping through the columns and creating a distribution plot for each one
for i, colu_val in enumerate(colu_names):
    sns.distplot(feats_df[colu_val], hist=True, ax=ax[i])
    ax[i].set_title('Freq distr '+colu_val, fontsize=10)
    ax[i].set_xlabel(colu_val, fontsize=8)
    ax[i].set_ylabel('Count', fontsize=8)
plt.show()
```

#### Boxplot Analysis: Speechiness, Danceability, Key

```python
sns.set(style='whitegrid')
ax = sns.boxplot(x=feats_df['speechiness'])
```

```python
sns.set(style='whitegrid')
ax = sns.boxplot(x=feats_df['danceability'], color='#fec508')
```

```python
sns.set(style='whitegrid')
ax = sns.boxplot(x=feats_df['key'], color='#ED4639')
```
